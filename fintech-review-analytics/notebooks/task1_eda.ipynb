```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.feature_extraction.text import TfidfVectorizer

# --- 1. Load the news data ---
df = pd.read_csv('data/raw/news.csv')
print(df.head())
print(df.info())

# --- 2. Descriptive Statistics ---
df['headline_length'] = df['headline'].str.len()
print(df['headline_length'].describe())          # headline character stats

publisher_counts = df['publisher'].value_counts()
print(publisher_counts.head(10))                 # top publishers

# --- 3. Text Analysis (keywords/topics) ---
tfidf = TfidfVectorizer(max_features=50, stop_words='english')
tfidf.fit(df['headline'])
print(tfidf.get_feature_names_out())             # top keywords

# --- 4. Time Series Analysis ---
df['date'] = pd.to_datetime(df['date'], utc=True)
df['date_only'] = df['date'].dt.date
articles_per_day = df.groupby('date_only').size()

plt.figure(figsize=(12, 4))
articles_per_day.plot()
plt.title('News Article Volume Over Time')
plt.savefig('notebooks/plots/news_volume.png')
plt.show()

# Publishing hour analysis
df['hour'] = df['date'].dt.hour
df['hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Articles by Hour of Day')
plt.show()

# --- 5. Publisher Analysis ---
# Extract domain if publisher is email
df['domain'] = df['publisher'].str.extract(r'@(.+)')
print(df['domain'].value_counts().head(10))
```
