```python
import pandas as pd
import talib
import matplotlib.pyplot as plt

# --- 1. Load stock price data ---
df = pd.read_csv('data/raw/AAPL_stock.csv', parse_dates=['Date'])
df.sort_values('Date', inplace=True)
df.dropna(inplace=True)

close = df['Close'].values
high  = df['High'].values
low   = df['Low'].values

# --- 2. Compute Technical Indicators ---
# Moving Averages
df['SMA_20'] = talib.SMA(close, timeperiod=20)
df['SMA_50'] = talib.SMA(close, timeperiod=50)
df['EMA_20'] = talib.EMA(close, timeperiod=20)

# RSI
df['RSI'] = talib.RSI(close, timeperiod=14)

# MACD
df['MACD'], df['MACD_signal'], df['MACD_hist'] = talib.MACD(close)

# --- 3. Visualize ---
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Price + MAs
axes[0].plot(df['Date'], df['Close'], label='Close')
axes[0].plot(df['Date'], df['SMA_20'], label='SMA 20')
axes[0].plot(df['Date'], df['SMA_50'], label='SMA 50')
axes[0].legend(); axes[0].set_title('Price with Moving Averages')

# RSI
axes[1].plot(df['Date'], df['RSI'], color='purple')
axes[1].axhline(70, color='red', linestyle='--', label='Overbought')
axes[1].axhline(30, color='green', linestyle='--', label='Oversold')
axes[1].legend(); axes[1].set_title('RSI')

# MACD
axes[2].plot(df['Date'], df['MACD'], label='MACD')
axes[2].plot(df['Date'], df['MACD_signal'], label='Signal')
axes[2].bar(df['Date'], df['MACD_hist'], alpha=0.3)
axes[2].legend(); axes[2].set_title('MACD')

plt.tight_layout()
plt.savefig('notebooks/plots/technical_indicators.png')
plt.show()
```
